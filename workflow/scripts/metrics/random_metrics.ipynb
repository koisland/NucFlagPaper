```python
import polars as pl
```

```python
df = pl.read_csv(
    "/project/logsdon_shared/projects/Keith/NucFlagPaper/workflow/scripts/metrics/out.csv",
    null_values=["None"],
).with_columns(is_small=pl.col("len").is_in([1, 2, 10, 100]))
df
```

```python
df_summ = (
    df.group_by(["dtype", "tool", "is_small"])
    .agg(
        pl.col("precision").mean(),
        pl.col("recall").mean(),
        pl.col("f1").mean(),
    )
    .sort("tool", "dtype")
)
df_summ
```

```python
df_summ.write_csv("summ.csv")
```
