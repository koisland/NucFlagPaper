Code to stratify simulated error size metrics

```python
# See merge_misasim.sh
df = pl.read_csv(
    "/project/logsdon_shared/projects/Keith/NucFlagPaper/workflow/scripts/metrics/out.csv",
    null_values=["None"],
).with_columns(is_small=pl.col("len").is_in([1, 2, 10, 100]))
df
```

```python
df_summ = (
    df.group_by(["dtype", "tool", "is_small"])
    .agg(
        pl.col("precision").mean(),
        pl.col("recall").mean(),
        pl.col("f1").mean(),
    )
    .sort("tool", "dtype")
)
df_summ
```

```python
df_summ.write_csv("summ.csv")
```

Code to aggregate homopolymer error counts

```python
import os
import glob

files = glob.glob(
    "/project/logsdon_shared/projects/Keith/NucFlagPaper/results/curated/homopolymers/*_bin_cnts.tsv"
)
dfs_hp = []
for file in files:
    sample = os.path.basename(file).replace("_bin_cnts.tsv", "")
    df_hp = (
        pl.read_csv(
            file,
            separator="\t",
            has_header=False,
            new_columns=["nt", "length", "count"],
        )
        .group_by(["nt"])
        .agg(pl.col("count").sum())
        .sort(by="nt")
        .with_columns(sample=pl.lit(sample))
    )
    dfs_hp.append(df_hp)
df_all_hp = pl.concat(dfs_hp).sort(by="nt")
df_all_hp.write_csv("out_hp_hg002.csv")
```

```python
files = glob.glob(
    "/project/logsdon_shared/projects/Keith/NucFlagPaper/results/assembly/homopolymers/*_bin_cnts.tsv"
)
dfs_hp = []
for file in files:
    sample = os.path.basename(file).replace("_bin_cnts.tsv", "")
    df_hp = (
        pl.read_csv(
            file,
            separator="\t",
            has_header=False,
            new_columns=["nt", "length", "count"],
        )
        .group_by(["nt"])
        .agg(pl.col("count").sum())
        .sort(by="nt")
        .with_columns(sample=pl.lit(sample))
    )
    dfs_hp.append(df_hp)
df_all_hp = pl.concat(dfs_hp).sort(by="nt")
df_all_hp.write_csv("out_hp_chm13.csv")
```
